```python
import pandas as pd
x = pd.read_csv('x.csv')
y = pd.read_csv('y.csv')
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
x = np.array(x['MW'])
#x is the input
X = np.array(x)
X = X.reshape(-1,1) #Change to the MinMaxScaler input format
minmax = MinMaxScaler().fit(X.astype('float32')) 
X_z = minmax.transform(X.astype('float32'))
```

```python
#y is the output
y=np.array(y['MW'])
y = np.array(y)
y = y.reshape(-1,1)
#Use the Minmaxscaler to normalize data
#LSTM prefers normalized data. e.g. from 0-1
minmax = MinMaxScaler().fit(y.astype('float32')) 
yy = minmax.transform(y.astype('float32'))
```

```python
#LSTM packages
import tensorflow as tf
from tensorflow.keras.models import Sequential
from keras.layers import LSTM, Bidirectional, Dense, Dropout
from tensorflow.keras.layers import LSTM, Dense, RepeatVector, TimeDistributed
from sklearn.preprocessing import MinMaxScaler
from keras.callbacks import EarlyStopping
from keras.layers import Dropout, BatchNormalization
from tensorflow import keras
from tensorflow.keras.layers import Layer
from tensorflow.keras.regularizers import l2
from tensorflow.keras.initializers import Initializer
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

Create a LSTM model input as a list of time series, with a single layer, to produce condensed output. In other words, each time series creates a scaler as output.

```python
# the difference between the last element of input and the output
lag = 1
sequences = X_z

#size: the length of each time series from x

def create(size, lag, sequences,yy):
    XX = []
    y = []
    for i in range(len(sequences)):
        end_ix = i + size
        out_end_ix = end_ix + lag
        
        if out_end_ix >= len(sequences):
            break

        if i+ size + lag >= len(yy):
            break
        start_ix = i
        out_loc = i+ size + lag
        print(start_ix, end_ix, out_loc)
        seq_x = sequences[start_ix: end_ix]
        XX.append(seq_x)



        seq_y = yy[out_loc]
        y.append(seq_y)

# LSTM requries a 3D-array so that I modify my input to three dimensional.
    XX = np.array(XX)
    y = np.array(y)
    return XX, y


size = 48
X,YYY = create(size, lag, sequences,yy)
```

```python
#parameters for LSTM model
SEED = 1
neurons=32
drop=.05
warnings.filterwarnings('ignore')
warnings.filterwarnings('ignore', category=UserWarning, module='absl')

keras.utils.set_random_seed(SEED)

#Create a single layer LSTM model and return_sequences equals to false, which creates a seq2num model
#units,dropout are parameters can be tuned.
model = Sequential()
model.add(LSTM(units=4, return_sequences=False, input_shape=(X.shape[1], X.shape[2]), dropout=drop))
#sigmoid is the most useful activation function for input between 0-1. 
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01), loss="mean_absolute_error")

#Create a early stop limit to avoid overfitting.
es = EarlyStopping(
  monitor="val_loss", mode="min", verbose=0, patience=10, restore_best_weights=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, YYY, test_size=0.2, random_state=42)
#Split train-test set and fit the model

#validation_split: cross-validation to aviod overfitting.
#epoch, batch_size are parameters
model.fit(X_train,
  y_train,
  validation_split=0.2,
  epochs=1000,
  batch_size=32,
  verbose=2,
  callbacks=[es],
)
```

I use two criterias for my result: First, the regular RMSE. Second, the percentage of predicted values fell into 20% error. In this example, it is 44%.

```python
zz = model.predict(X_test)
o = minmax.inverse_transform(zz)
o2 = minmax.inverse_transform(y_test)
len(o)

a = (o-o2)/o2
count = 0 
for i in a:
    if -0.2 < i < 0.2:
        count += 1
print(count/len(o))
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

```

Now it is hyperparameter tuning. I choose 6 parameters.  \
LAG: the time difference between the end of input and the output. For example, if I use whole day info on 03/01/2024 (24 inputs) to predict 03/02/2024 06:00am power generation. In this case, LAG parameter is 6. \
neurons: numbers of neurons of the algorithm \
drop: drop_out rate \
units: the number of units of the algorithm \
lr: learning_rate \
size: the timeframe of the input. For example, if I use infos from 03/01/2024 to 03/02/2024, SIZE parameter is 48. \
bs: batch_size

```python
#create a hyper parameter function that help to tune easily.
rrs = []
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
def hyper(lag, neurons, drop, units,size,lr,bs, rrs): #Store results in rrs
    lag = lag
    sequences = X_z
    y_z = yy
    def create(size, lag, sequences, y_z):
        XX = []
        y = []
        for i in range(len(sequences)):
            end_ix = i + size
            out_end_ix = end_ix + lag

            if out_end_ix >= len(sequences):
                break
            if i+ size + lag >= len(y_z):
                break

            start_ix = i
            out_loc = i+ size + lag
            seq_x = sequences[start_ix: end_ix]
            XX.append(seq_x)


            seq_y = y_z[out_loc]
            y.append(seq_y)

        XX = np.array(XX)
        y = np.array(y)
        return XX, y

    XX,YY = create(size, lag, sequences,y_z)
    SEED = 1
    neurons=neurons
    drop=drop
    
    model = Sequential()
    model.add(LSTM(units=units, return_sequences=False, input_shape=(XX.shape[1], XX.shape[2]), dropout=drop))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr), loss="mean_absolute_error")
    es = EarlyStopping(
        monitor="val_loss", mode="min", verbose=0, patience=5, restore_best_weights=True)

    X_train, X_test, y_train, y_test = train_test_split(XX, YY, test_size=0.2, random_state=42)
    model.fit(X_train,
      y_train,
      validation_split=0.2,
      epochs=1000,
      batch_size=bs,
      verbose=2,
      callbacks=[es],  
)
    
    zz = model.predict(X_test)
    o = minmax.inverse_transform(zz)
    o2 = minmax.inverse_transform(y_test)
    len(o)

    a = (o-o2)/o2
    count = 0 
    for i in a:
        if -0.2 < i < 0.2:
            count += 1
    res = count/len(o)
    rr = rmse(o, o2)
    print(rr)
    
    rrs.append([lag,neurons, drop, units, size,lr, bs, rr,res])

    print(f'start lag = {lag}, neurons = {neurons}, drop_rate = {drop}, units = {units}, x window size = {size}, learning_rate = {lr}, batch_size = {bs}')
    print(f'The percentage of error falls into 20% is {res}')
```

I want to predict the result for the following 24 data. By the tuning result, I need the window size of 48 with lag 1.

```python
ipt = pd.read_csv('ipt.csv')
ipt = ipt['MW']
ipt

ipt = np.array(ipt)
ipt = ipt.reshape(-1,1)
minmax = MinMaxScaler().fit(ipt.astype('float32')) # Close index
ipt = minmax.transform(ipt.astype('float32'))
ipt.shape
print(ipt)


#create dataset
lg = 1
sequences = ipt
def create(size, lag, sequences):
    XX = []
    for i in range(24):
        end_ix = i + size
        out_end_ix = end_ix + lag
        start_ix = i
        out_loc = i+ size + lag
        seq_x = sequences[start_ix: end_ix]
        XX.append(seq_x)

        
    XX = np.array(XX)

    return XX

sz = 48
XXX = create(sz, lg, sequences)
XXX
```

```python
#create the model for predicting result
def predict(lag, neurons, drop, units,size,lr,bs,XXX):
    lag = lag
    sequences = X_z
    y_z = yy
    def create(size, lag, sequences, y_z):
        XX = []
        y = []
        for i in range(len(sequences)):
            end_ix = i + size
            out_end_ix = end_ix + lag

            if out_end_ix >= len(sequences):
                break
            if i+ size + lag >= len(y_z):
                break

            start_ix = i
            out_loc = i+ size + lag
            seq_x = sequences[start_ix: end_ix]
            XX.append(seq_x)


            seq_y = y_z[out_loc]
            y.append(seq_y)

        XX = np.array(XX)
        y = np.array(y)
        return XX, y

    XX,YY = create(size, lag, sequences,y_z)
    SEED = 1
    neurons=neurons
    drop=drop
    
    model = Sequential()
    model.add(LSTM(units=units, return_sequences=False, input_shape=(XX.shape[1], XX.shape[2]), dropout=drop))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr), loss="mean_absolute_error")
    es = EarlyStopping(
        monitor="val_loss", mode="min", verbose=0, patience=5, restore_best_weights=True)

  
    
    model.fit(XX,
      YY,
      validation_split=0.2,
      epochs=1000,
      batch_size=bs,
      verbose=0,
      callbacks=[es],  
)
    
    
    zz = model.predict(XXX)
    o = minmax.inverse_transform(zz)
    return o
```
